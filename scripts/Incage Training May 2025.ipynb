```python
import matplotlib.pyplot as plt
import numpy as np

#Trial and reward pulse functions

def plot_trial_time(df, label='Dataset'):
    """
    Plots trial start times with accuracy indicators and displays accuracy percentages as an overlay bar plot.

    Parameters:
    - df (pd.DataFrame): DataFrame containing at least 'time' and 'acc' columns.
    - label (str): Optional label to identify the dataset in the plot title.

    Returns:
    - None
    """
    # Check if required columns exist
    if not {'time', 'acc'}.issubset(df.columns):
        raise ValueError("DataFrame must contain 'time' and 'acc' columns.")

    # Convert trial start times to hours
    trial_start_times_hours = df['time'] / 3600  # Convert seconds to hours
    trial_numbers = range(1, len(trial_start_times_hours) + 1)  # Accumulated trial numbers

    # Calculate overall accuracy percentage
    total_trials = len(df)
    if total_trials == 0:
        overall_accuracy = 0
    else:
        overall_accuracy = (df['acc'].sum() / total_trials) * 100

    # Define time windows (Extended to 6 hours)
    time_windows = {
        '0-1hr': (0, 1),
        '1-2hr': (1, 2),
        '2-3hr': (2, 3),
        '3-4hr': (3, 4),
        '4-5hr': (4, 5),
        '5-6hr': (5, 6)
    }

    accuracies = {}
    for window, (start, end) in time_windows.items():
        mask = (trial_start_times_hours >= start) & (trial_start_times_hours < end)
        window_trials = df[mask]
        num_window_trials = len(window_trials)
        if num_window_trials == 0:
            accuracies[window] = 0
        else:
            accuracies[window] = (window_trials['acc'].sum() / num_window_trials) * 100

    # Create a figure for the plot
    fig, ax1 = plt.subplots(figsize=(8, 5))

    # Plot green dots for correct trials and red dots for incorrect trials
    correct_trials = trial_start_times_hours[df['acc'] == 1]
    correct_numbers = [trial_numbers[i] for i in df.index[df['acc'] == 1]]
    incorrect_trials = trial_start_times_hours[df['acc'] == 0]
    incorrect_numbers = [trial_numbers[i] for i in df.index[df['acc'] == 0]]

    # Overlay bar plot for accuracies first
    ax1.set_ylabel('Accuracy (%)', color='teal')
    ax1.tick_params(axis='y', labelcolor='teal')
    accuracy_values = list(accuracies.values())
    accuracy_labels = list(accuracies.keys())
    x_positions = range(len(accuracy_labels))
    ax1.bar(x_positions, accuracy_values, alpha=0.5, edgecolor='black', color='azure', align='edge', width=0.5)
    ax1.set_xticks(x_positions)
    ax1.set_xticklabels(accuracy_labels)
    ax1.set_ylim(0, 100)

    # Plot green dots for correct trials and red dots for incorrect trials
    correct_trials = trial_start_times_hours[df['acc'] == 1]
    correct_numbers = [trial_numbers[i] for i in df.index[df['acc'] == 1]]
    incorrect_trials = trial_start_times_hours[df['acc'] == 0]
    incorrect_numbers = [trial_numbers[i] for i in df.index[df['acc'] == 0]]

    ax2 = ax1.twinx()
    ax2.scatter(correct_trials, correct_numbers, color='green', marker='x', s=0.1, label=f'Correct (Overall:{overall_accuracy:.2f}%)')
    ax2.scatter(incorrect_trials, incorrect_numbers, color='red', marker='x', s=0.1, label='Incorrect ')

    # Set labels and title
    ax2.set_xlabel('Time (hours)', color='green')
    ax2.set_ylabel('Accumulated Trial Number', color='green')
    ax2.tick_params(axis='y', labelcolor='green')
    total_pulses = df['pulses'].sum()  # Sum of all pulse values
    total_volume = total_pulses * 0.3  # Total volume in mL
    ax2.set_title(f'Trial Start Times with Accuracy for {label} (Total: {len(df)} trials, {total_pulses} pulses, {total_volume:.1f} mL)')
    ax2.legend(loc='upper right')
    ax2.grid(True)

    # Tight layout and show
    plt.tight_layout()
    plt.show()

def reward_amount(df):
    counts = df['pulses'].value_counts().sort_index()
    num_twos = counts.get(2, 0)
    total_count = len(df['pulses'])
    total_pulses = df['pulses'].sum()
    total_volume = total_pulses * 0.3  # mL
    percent_twos = (num_twos / total_count) * 100 if total_count > 0 else 0
    water_left = max(0, 850 - total_volume)  # in mL

    # --- Plotting ---
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    # Histogram of pulse counts
    bars = ax1.bar(counts.index, counts.values, tick_label=counts.index)
    for i, count in enumerate(counts.values):
        percent = (count / total_count) * 100
        ax1.text(counts.index[i], count + 0.5, f"{percent:.1f}%", ha='center', va='bottom', fontsize=10)
    ax1.set_xlabel('Pulse Value')
    ax1.set_ylabel('Count')
    ax1.set_title(f'Pulse Histogram\nTotal: {total_count} trials, {total_pulses} pulses, {total_volume:.1f} mL')

    # Water bottle level in mL
    ax2.bar(0, 900, color='lightgray', width=0.4, edgecolor='black', label='Bottle Capacity')
    ax2.bar(0, water_left, color='blue', width=0.4, edgecolor='black', label='Water Left')

    ax2.axhline(900, color='black', linestyle='--', linewidth=1, label='Full (900 mL)')
    ax2.set_ylim(0, 950)
    ax2.set_xlim(-1, 1)
    ax2.set_ylabel('Water (mL)')
    ax2.set_xticks([])
    ax2.set_title(f'Water Left: {water_left:.1f} mL')
    ax2.legend()

    plt.tight_layout()
    plt.show()

def plot_consumption_and_trials_across_days(datasets, labels):
    """
    Plots the consumption level (total volume in mL) and the number of trials across multiple days.

    Parameters:
    - datasets (list of pd.DataFrame): List of DataFrames containing 'pulses' and 'trial' columns.
    - labels (list of str): List of labels corresponding to each dataset.

    Returns:
    - None
    """
    if len(datasets) != len(labels):
        raise ValueError("The number of datasets must match the number of labels.")

    total_volumes = []
    total_trials = []
    for df in datasets:
        total_pulses = df['pulses'].sum()  # Sum of all pulse values
        total_volume = total_pulses * 0.3  # Total volume in mL
        total_volumes.append(total_volume)
        total_trials.append(len(df))  # Count the number of trials

    # Create a figure with two y-axes
    fig, ax1 = plt.subplots(figsize=(10, 6))

    # Plot consumption levels
    ax1.plot(labels, total_volumes, color='blue', marker='o',label='Total Consumption (mL)')
    ax1.set_ylabel('Total Consumption (mL)', color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')
    ax1.set_xlabel('Day')
    ax1.set_xticks(range(len(labels)))
    ax1.set_xticklabels(labels, rotation=45, ha='right')

    # Create a second y-axis for the number of trials
    ax2 = ax1.twinx()
    ax2.plot(labels, total_trials, color='orange', marker='o', label='Number of Trials')
    ax2.set_ylabel('Number of Trials', color='orange')
    ax2.tick_params(axis='y', labelcolor='orange')

    # Add title and grid
    plt.title('Consumption Level and Number of Trials Across Days')
    ax1.grid(axis='y', linestyle='--', alpha=0.7)

    # Add legends
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')

    plt.tight_layout()
    plt.show()

def plot_reward_volume_vs_trial_count(datasets, labels):
    """
    Plots total reward volume (in mL) divided by total trial count across multiple days.

    Parameters:
    - datasets (list of pd.DataFrame): List of DataFrames containing 'pulses' and 'trial' columns.
    - labels (list of str): List of labels corresponding to each dataset.

    Returns:
    - None
    """
    if len(datasets) != len(labels):
        raise ValueError("The number of datasets must match the number of labels.")

    reward_volumes_per_trial = []
    for df in datasets:
        total_pulses = df['pulses'].sum()  # Sum of all pulse values
        total_volume = total_pulses * 0.3  # Total volume in mL
        total_trials = len(df)  # Count the number of trials
        reward_volumes_per_trial.append(total_volume / total_trials if total_trials > 0 else 0)

    # Plot reward volume per trial across days
    plt.figure(figsize=(12, 6))
    plt.plot(labels, reward_volumes_per_trial, marker='o', linestyle='-', color='blue', alpha=0.7)
    plt.xlabel('Day')
    plt.ylabel('Reward Volume per Trial (mL)')
    plt.title('Reward Volume per Trial Across Days')
    plt.grid(axis='both', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
# Accuracy functions

def accuracy_moving_window(df, window_size=50, step_size=2):

    # Calculate moving average accuracy for both trial types within the window
    moving_target_accuracy = []
    moving_catch_accuracy = []
    false_alarm_rate = []

    for i in range(0, len(df) - window_size + 1, step_size):
        window = df.iloc[i:i + window_size]
        target_trials = window[window['trial_type'] == 'target']
        catch_trials = window[window['trial_type'] == 'catch']
        incorrect_catch_trials = window[(window['trial_type'] == 'catch') & (window['acc'] == 0)]
        early_release_target_trials = window[(window['trial_type'] == 'target') & (window['acc'] == 0) & 
                                         (window['rt'].str.contains('cti|fix|bar', na=False))|(pd.to_numeric(window['rt'], errors='coerce') < 0.2)]
        total_trials_in_window = len(window)

        if len(target_trials) > 0:
            moving_target_accuracy.append(target_trials['acc'].mean() * 100)
        else:
            moving_target_accuracy.append(None)

        if len(catch_trials) > 0:
            moving_catch_accuracy.append(catch_trials['acc'].mean() * 100)
        else:
            moving_catch_accuracy.append(None)

        if total_trials_in_window > 0:
            false_alarm_rate.append((len(incorrect_catch_trials) + len(early_release_target_trials)) / total_trials_in_window * 100)
        else:
            false_alarm_rate.append(None)

    # Plot the moving average accuracies
    target_accuracy = df[df['trial_type'] == 'target']['acc'].mean() * 100
    catch_accuracy = df[df['trial_type'] == 'catch']['acc'].mean() * 100
    overall_target_early_release = len(df[(df['trial_type'] == 'target') & (df['acc'] == 0) & ((df['rt'].str.contains('cti|fix|bar', na=False)) | (pd.to_numeric(df['rt'], errors='coerce') < 0.2))])
    overall_incorrect_catch= len(df[(df['trial_type'] == 'catch') & (df['acc'] == 0)])
    overall_fa_rate= ((overall_target_early_release+overall_incorrect_catch)/len(df))*100
    
    plt.figure(figsize=(10, 5)) 
    plt.plot(range(0, len(moving_target_accuracy) * step_size, step_size), moving_target_accuracy, label=f'Moving Target Trial Accuracy (Overall: {target_accuracy:.2f}%)', color='green', alpha=0.7)
    plt.plot(range(0, len(moving_catch_accuracy) * step_size, step_size), moving_catch_accuracy, label=f'Moving Catch Trial Accuracy (Overall: {catch_accuracy:.2f}%)', color='red', alpha=0.7)
    plt.plot(range(0, len(false_alarm_rate) * step_size, step_size), false_alarm_rate, label=f'False Alarm Rate (Overall: {overall_fa_rate:.2f}%)', color='blue', alpha=0.7)
    plt.ylim(0, 101)
    plt.yticks(np.arange(0, 101, 10))
    plt.xlabel('Window Start Index')
    plt.ylabel('Accuracy (%)')
    plt.title(f'Moving Window Accuracy (Window Size: {window_size}, Step Size: {step_size})')
    plt.legend()
    plt.grid(visible=True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
    plt.show()

    
    
def plot_accuracy_multiple_datasets(datasets, labels, window_size=50, step_size=2): 
    """ 
    Plots moving window accuracy for multiple datasets in a continuous plot without gaps between datasets.

    Parameters:
    - datasets (list of pd.DataFrame): List of DataFrames containing 'trial_type' and 'acc' columns.
    - labels (list of str): List of labels corresponding to each dataset.
    - window_size (int): Size of the moving window.
    - step_size (int): Step size for the moving window.

    Returns:
    - None
    """
    if len(datasets) != len(labels):
        raise ValueError("The number of datasets must match the number of labels.")

    plt.figure(figsize=(12, 6))
    current_offset = 0

    for i, (df, label) in enumerate(zip(datasets, labels)):
        # Calculate moving average accuracy for both trial types within the window
        moving_target_accuracy = []
        moving_catch_accuracy = []
        false_alarm_rate = []

        for j in range(0, len(df) - window_size + 1, step_size):
            window = df.iloc[j:j + window_size]
            target_trials = window[window['trial_type'] == 'target']
            catch_trials = window[window['trial_type'] == 'catch']
            incorrect_catch_trials = window[(window['trial_type'] == 'catch') & (window['acc'] == 0)]
            early_release_target_trials = window[(window['trial_type'] == 'target') & (window['acc'] == 0) & 
                                            (window['rt'].str.contains('cti|fix|bar', na=False))]
            total_trials_in_window = len(window)

            if len(target_trials) > 0:
                moving_target_accuracy.append(target_trials['acc'].mean() * 100)
            else:
                moving_target_accuracy.append(None)

            if len(catch_trials) > 0:
                moving_catch_accuracy.append(catch_trials['acc'].mean() * 100)
            else:
                moving_catch_accuracy.append(None)

            if total_trials_in_window > 0:
                false_alarm_rate.append((len(incorrect_catch_trials) + len(early_release_target_trials)) / total_trials_in_window * 100)
            else:
                false_alarm_rate.append(None)

        # Plot the moving average accuracies
        x_values = range(current_offset, current_offset + len(moving_target_accuracy) * step_size, step_size)
        plt.plot(x_values, moving_target_accuracy, color='green', alpha=0.7, label=f'Target Trial Accuracy' if i == 0 else None)
        plt.plot(x_values, moving_catch_accuracy, color='red', alpha=0.7, label=f'Catch Trial Accuracy' if i == 0 else None)
        plt.plot(x_values, false_alarm_rate, label=f'False Alarm Rate' if i == 0 else None, color='blue', alpha=0.7)

        # Wrap long labels to fit into multiple lines
        wrapped_label = label.replace('\\', '\n')
        # Add a vertical line to delineate datasets, with text annotations at the top
        plt.axvline(x=current_offset, color='cyan', linestyle='--', alpha=1)
        plt.text(current_offset, 15, wrapped_label, rotation=20, horizontalalignment='center', fontsize=8, weight='bold', verticalalignment='bottom', transform=plt.gca().transData)

        # Update offset for the next dataset
        current_offset += len(moving_target_accuracy) * step_size

    plt.ylim(0, 100)
    plt.xlabel('Trial Index (Continuous)')
    plt.ylabel('Accuracy (%)')
    plt.title(f'Moving Window Accuracy Across Multiple Datasets (Window Size: {window_size}, Step Size: {step_size})', pad=40, loc='center', y=-0.25)
    plt.legend()
    plt.grid(visible=True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
    plt.tight_layout()
    plt.show()

def calculate_stats(df):
    """
    Calculates and prints statistics for the given DataFrame.

    Parameters:
    - df (pd.DataFrame): DataFrame containing at least 'time' and 'acc' columns.

    Returns:
    - None
    """
    if not {'time', 'acc','trial_type'}.issubset(df.columns):
        raise ValueError("DataFrame must contain 'trial_type', 'time', and 'acc' columns.")

    target_trials=df[df['trial_type']=='target']
    catch_trials=df[df['trial_type']=='catch']

    # Calculate statistics
    target_accuracy = target_trials['acc'].mean() * 100
    catch_accuracy = catch_trials['acc'].mean() * 100

    overall_target_early_release = len(df[(df['trial_type'] == 'target') & (df['acc'] == 0) & ((df['rt'].str.contains('cti|fix|bar', na=False)) | (pd.to_numeric(df['rt'], errors='coerce') < 0.2))])
    overall_incorrect_catch= len(df[(df['trial_type'] == 'catch') & (df['acc'] == 0)])
    overall_fa_rate= ((overall_target_early_release+overall_incorrect_catch)/len(df))*100
    
    # Calculate counts for incorrect and correct trials
    target_correct = len(target_trials[target_trials['acc'] == 1])
    catch_correct = len(catch_trials[catch_trials['acc'] == 1])

    # Print statistics with fractions
    print(f"Total Trials: {len(df)}")
    print(f"Target Trial Accuracy (hits/all target trials): {target_accuracy:.2f}% ({target_correct}/{len(target_trials)})")
    print(f"Catch Trial Accuracy: {catch_accuracy:.2f}% ({catch_correct}/{len(catch_trials)})")
    print(f"False Alarm Rate: {overall_fa_rate:.2f}% (({overall_target_early_release}+{overall_incorrect_catch})/{len(df)})")

def calculate_stats_table(datasets, labels):
    """
    Calculates statistics for multiple datasets and outputs them as a table.

    Parameters:
    - datasets (list of pd.DataFrame): List of DataFrames containing 'trial_type', 'time', and 'acc' columns.
    - labels (list of str): List of labels corresponding to each dataset.

    Returns:
    - pd.DataFrame: A DataFrame containing the calculated statistics for each dataset.
    """
    if len(datasets) != len(labels):
        raise ValueError("The number of datasets must match the number of labels.")

    stats = []

    for df, label in zip(datasets, labels):
        if not {'time', 'acc', 'trial_type'}.issubset(df.columns):
            raise ValueError("Each DataFrame must contain 'trial_type', 'time', and 'acc' columns.")

        target_trials = df[df['trial_type'] == 'target']
        catch_trials = df[df['trial_type'] == 'catch']

        # Calculate statistics
        target_accuracy = target_trials['acc'].mean() * 100
        catch_accuracy = catch_trials['acc'].mean() * 100

        overall_target_early_release = len(df[(df['trial_type'] == 'target') & (df['acc'] == 0) & (df['rt'].str.contains('cti|fix|bar', na=False)| (pd.to_numeric(df['rt'], errors='coerce') < 0.2))])
        overall_incorrect_catch = len(df[(df['trial_type'] == 'catch') & (df['acc'] == 0)])
        overall_fa_rate = ((overall_target_early_release + overall_incorrect_catch) / len(df)) * 100

        # Calculate counts for incorrect and correct trials
        target_correct = len(target_trials[target_trials['acc'] == 1])
        catch_correct = len(catch_trials[catch_trials['acc'] == 1])

        stats.append({
            'Dataset': label,
            'Total Trials': len(df),
            'Target Accuracy (%)': f"{target_accuracy:.2f} ({target_correct}/{len(target_trials)})",
            'Catch Accuracy (%)': f"{catch_accuracy:.2f} ({catch_correct}/{len(catch_trials)})",
            'False Alarm Rate (%)': f"{overall_fa_rate:.2f} (({overall_target_early_release}+{overall_incorrect_catch})/{len(df)})"
        })

    stats_df = pd.DataFrame(stats)
    
    # Plot the table
    fig, ax = plt.subplots(figsize=(10, len(stats_df) * 0.05))  # Adjust height to reduce empty space
    ax.axis('off')
    table = ax.table(cellText=stats_df.values, colLabels=stats_df.columns, cellLoc='center', loc='center')
    table.auto_set_font_size(True)
    table.auto_set_column_width(col=list(range(len(stats_df.columns))))
    plt.show()

def plot_stats_table(datasets, labels):
    """
    Plots a statistics table as a bar chart for multiple datasets.

    Parameters:
    - datasets (list of pd.DataFrame): List of DataFrames containing 'trial_type', 'acc', and other relevant columns.
    - labels (list of str): List of labels corresponding to each dataset.

    Returns:
    - None
    """
    if len(datasets) != len(labels):
        raise ValueError("The number of datasets must match the number of labels.")

    stats = []

    for df, label in zip(datasets, labels):
        if not {'trial_type', 'acc'}.issubset(df.columns):
            raise ValueError("Each DataFrame must contain 'trial_type' and 'acc' columns.")

        target_trials = df[df['trial_type'] == 'target']
        catch_trials = df[df['trial_type'] == 'catch']

        # Calculate statistics
        target_accuracy = target_trials['acc'].mean() * 100 if len(target_trials) > 0 else 0
        catch_accuracy = catch_trials['acc'].mean() * 100 if len(catch_trials) > 0 else 0

        overall_target_early_release = len(df[(df['trial_type'] == 'target') & (df['acc'] == 0) & 
                                              (df['rt'].str.contains('cti|fix|bar', na=False))])
        overall_incorrect_catch = len(df[(df['trial_type'] == 'catch') & (df['acc'] == 0)])
        overall_fa_rate = ((overall_target_early_release + overall_incorrect_catch) / len(df)) * 100 if len(df) > 0 else 0

        stats.append({
            'Dataset': label,
            'Target Accuracy (%)': target_accuracy,
            'Catch Accuracy (%)': catch_accuracy,
            'False Alarm Rate (%)': overall_fa_rate
        })

    stats_df = pd.DataFrame(stats)

    # Plot the statistics table as a bar chart
    fig, ax = plt.subplots(figsize=(12, 6))

    # Extract datasets, target accuracy, catch accuracy, and false alarm rate
    datasets = stats_df['Dataset']
    target_accuracy = stats_df['Target Accuracy (%)']
    catch_accuracy = stats_df['Catch Accuracy (%)']
    false_alarm_rate = stats_df['False Alarm Rate (%)']
    
    # Bar width and positions
    x = np.arange(len(datasets))

    # Plot lines for each metric
    ax.plot(x, target_accuracy, label='Target Accuracy (%)', color='green', alpha=0.7, marker='o', linestyle='-')
    ax.plot(x, catch_accuracy, label='Catch Accuracy (%)', color='red', alpha=0.7, marker='o', linestyle='-')
    ax.plot(x, false_alarm_rate, label='False Alarm Rate (%)', color='blue', alpha=0.7, marker='o', linestyle='-')

    # Add labels, title, and legend
    ax.set_xlabel('Dataset', fontsize=12)
    ax.set_ylabel('Percentage (%)', fontsize=12)
    ax.set_title('Performance Metrics by Dataset', fontsize=14)
    ax.set_xticks(x)
    ax.set_yticks(np.arange(0, 101, 10))
    ax.set_xticklabels(datasets, rotation=45, ha='right', fontsize=10)
    ax.legend()
    ax.grid(axis='y', linestyle='--', alpha=0.7)

    # Adjust layout and show plot
    plt.tight_layout()
    plt.show()

    return stats_df


# Reaction time functions

def plot_cti_vs_accuracy(df, label='Dataset',bins=30):
    # Filter for 'target' trials and remove NaNs
    df = df.loc[(df['trial_type'] == 'target') & df['cti'].notna() & df['acc'].notna()].copy()

    # Define CTI bins
    num_bins = bins
    bins = np.linspace(df['cti'].min(), df['cti'].max(), num_bins + 1)

    # Assign each CTI to a bin
    df['cti_bin'] = pd.cut(df['cti'], bins=bins, include_lowest=True)

    # Compute mean accuracy per CTI bin
    hit_rates = df.groupby('cti_bin', observed=False)['acc'].mean()
    bin_centers = [interval.mid for interval in hit_rates.index]

    # Plot
    plt.figure(figsize=(8, 5))
    plt.plot(bin_centers, hit_rates.values, marker='o')
    plt.xlabel('Cue-to-Target Interval (CTI)')
    plt.ylabel('Hit rate')
    plt.title(f'Hit rate vs CTI (Target Trials Only) {label}')
    plt.grid(True)
    plt.tight_layout()
    plt.show()


def plot_hit_rate_vs_cti(df):
    """
    Generates a plot of hit rate vs. CTI value.
    
    The function assumes that df has at least the following columns:
      - 'rt': reaction time values that can be numeric, NaN, or strings (e.g., 'nan', 'fix', 'bar', 'cti')
      - 'trial_type': should include at least 'target' for target trials.
      - 'cti': a numeric column representing the cue-target interval value.
    
    For target trials:
      - "Hits" are defined as those where the transformed RT is between 0 and 0.85.
      - "Misses" (Held Lever) are those with a transformed RT equal to 1.
      
    Hit rate is calculated as:
        hit_rate = hits / (hits + misses)
    """
    
    # Helper function to transform RT values
    def transform_val(val):
        # Treat NaN (or the string "nan") as no response (Held Lever: 1)
        if pd.isna(val) or (isinstance(val, str) and val.lower() == 'nan'):
            return 1
        # Special phase labels: 'fix', 'bar', or 'cti'
        if isinstance(val, str):
            if "fix" in val:
                return -0.3
            elif "bar" in val:
                return -0.2
            elif "cti" in val:
                return -0.1
        # Otherwise, attempt conversion to float
        try:
            return float(val)
        except (ValueError, TypeError):
            return np.nan

    # Work on a copy so the original DataFrame is not modified
    df = df.copy()
    df['rt_special'] = df['rt'].apply(transform_val)
    
    # Focus only on target trials
    target_df = df[df['trial_type'] == 'valid']
    
    # Group target trials by CTI value and compute hit rates
    hit_rates = {}
    for cti_val, group in target_df.groupby('cti'):
        # Hits: RT values between 0 and 0.85 (correct responses)
        hits = group[(group['rt_special'] > 0.2) & (group['rt_special'] < 0.8)].shape[0]
        # Misses: Held lever responses (rt_special == 1)
        misses = group[group['rt_special'] == 1].shape[0]
        total = hits + misses
        hit_rate = hits / total if total > 0 else np.nan
        hit_rates[cti_val] = hit_rate
    
    # Sort CTI values for plotting
    sorted_cti = sorted(hit_rates.keys())
    sorted_hit_rates = [hit_rates[cti] for cti in sorted_cti]
    
    # Create the plot
    plt.figure(figsize=(8, 6))
    plt.plot(sorted_cti, sorted_hit_rates, marker='o', linestyle='-')
    plt.xlabel('CTI Value')
    plt.ylabel('Hit Rate')
    plt.ylim(0,1)
    plt.title('Hit Rate vs. CTI Value')
    plt.grid(True)
    plt.tight_layout()
    plt.show()


def plot_cti_bins_vs_fa_rate(df, num_bins=20):
        """
        Plots FA rate vs CTI bins.

        Parameters:
        - df (pd.DataFrame): DataFrame containing 'trial_type', 'cti', and 'acc' columns.
        - num_bins (int): Number of bins for CTI.

        Returns:
        - None
        """
        # Filter for trials with valid CTI values
        filtered_df = df[df['cti'].notna()].copy()

        # Define CTI bins
        bins = np.linspace(filtered_df['cti'].min(), filtered_df['cti'].max(), num_bins + 1)

        # Assign each CTI to a bin
        filtered_df['cti_bin'] = pd.cut(filtered_df['cti'], bins=bins, include_lowest=True)

        # Calculate FA rate for each CTI bin
        fa_rates = []
        bin_centers = []
        for cti_bin, group in filtered_df.groupby('cti_bin', observed=False):
            overall_target_early_release = len(group[(group['trial_type'] == 'target') & (group['acc'] == 0) & ((group['rt'].str.contains('cti|fix|bar', na=False)) | (pd.to_numeric(group['rt'], errors='coerce') < 0.2))])
            overall_incorrect_catch = len(group[(group['trial_type'] == 'catch') & (group['acc'] == 0)])
            overall_fa_rate = ((overall_target_early_release + overall_incorrect_catch) / len(group)) * 100 if len(group) > 0 else 0
            fa_rates.append(overall_fa_rate)
            bin_centers.append(cti_bin.mid)

        # Plot FA rate vs CTI bins
        plt.figure(figsize=(8, 5))
        plt.plot(bin_centers, fa_rates, marker='o', color='blue', label='FA Rate')
        plt.xlabel('Cue-to-Target Interval (CTI)')
        plt.ylabel('False Alarm Rate (%)')
        plt.ylim(0,100)
        plt.title('FA Rate vs CTI Bins')
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()

def plot_cti_vs_acc_multiple_days(datasets, labels, num_bins=20):
    """
    Overlays multiple days of Accuracy vs CTI plots.

    Parameters:
    - datasets (list of pd.DataFrame): List of DataFrames containing 'trial_type', 'cti', and 'acc' columns.
    - labels (list of str): List of labels corresponding to each dataset.
    - num_bins (int): Number of bins for CTI.

    Returns:
    - None
    """
    if len(datasets) != len(labels):
        raise ValueError("The number of datasets must match the number of labels.")

    plt.figure(figsize=(10, 6))

    for df, label in zip(datasets, labels):
        # Filter for 'target' trials and remove NaNs
        filtered_df = df[(df['trial_type'] == 'target') & df['cti'].notna() & df['acc'].notna()].copy()

        # Define CTI bins
        bins = np.linspace(filtered_df['cti'].min(), filtered_df['cti'].max(), num_bins + 1)

        # Assign each CTI to a bin
        filtered_df['cti_bin'] = pd.cut(filtered_df['cti'], bins=bins, include_lowest=True)

        # Compute mean accuracy per CTI bin
        hit_rates = filtered_df.groupby('cti_bin', observed=False)['acc'].mean()
        bin_centers = [interval.mid for interval in hit_rates.index]

        # Plot
        plt.plot(bin_centers, hit_rates.values, label=label,marker='o')

    plt.xlabel('Cue-to-Target Interval (CTI)')
    plt.ylabel('Accuracy')
    plt.title('Accuracy vs CTI (Target Trials Only)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def plot_lever_release_period(df, label='Dataset'):
    """
    Plot lever release period for both target and catch trials.

    Categories:
    - Fixation: 'rt' column contains 'fix'
    - Bar: 'rt' column contains 'bar'
    - CTI: 'rt' column contains 'cti'
    - <0.2s in target: 'rt' value < 0.2
    - Target: 'rt' value >= 0.2
    - Post-target: 'rt' value is NaN
    """
    # Define categories
    categories = ['fixation', 'bar', 'cti', '<0.2s in target', 'target', 'post-target']
    catch_counts = dict.fromkeys(categories, 0)
    target_counts = dict.fromkeys(categories, 0)

    # Count lever lifts for each category
    catch_counts['fixation'] = len(df[(df['trial_type'] == 'catch') & (df['rt'].str.contains('fix', na=False))])
    target_counts['fixation'] = len(df[(df['trial_type'] == 'target') & (df['rt'].str.contains('fix', na=False))])

    catch_counts['bar'] = len(df[(df['trial_type'] == 'catch') & (df['rt'].str.contains('bar', na=False))])
    target_counts['bar'] = len(df[(df['trial_type'] == 'target') & (df['rt'].str.contains('bar', na=False))])

    catch_counts['cti'] = len(df[(df['trial_type'] == 'catch') & (df['rt'].str.contains('cti', na=False)|(df['rt'].str.contains(',nan', na=False)))])
    target_counts['cti'] = len(df[(df['trial_type'] == 'target') & (df['rt'].str.contains('cti', na=False))])

    catch_counts['<0.2s in target'] = len(df[(df['trial_type'] == 'catch') & (pd.to_numeric(df['rt'], errors='coerce') < 0.2)])
    target_counts['<0.2s in target'] = len(df[(df['trial_type'] == 'target') & (pd.to_numeric(df['rt'], errors='coerce') < 0.2)])

    catch_counts['target'] = len(df[(df['trial_type'] == 'catch') & (pd.to_numeric(df['rt'], errors='coerce') >= 0.2)])
    target_counts['target'] = len(df[(df['trial_type'] == 'target') & (pd.to_numeric(df['rt'], errors='coerce') >= 0.2)])

    catch_counts['post-target'] = len(df[(df['trial_type'] == 'catch') & (df['rt'].isna())])
    target_counts['post-target'] = len(df[(df['trial_type'] == 'target') & (df['rt'].isna())])

    # Prepare data for plotting
    data = pd.DataFrame({
        'Catch': [catch_counts[cat] for cat in categories],
        'Target': [target_counts[cat] for cat in categories]
    }, index=categories)

    # Calculate total trials for each category
    total_catch_trials = sum(catch_counts.values())
    total_target_trials = sum(target_counts.values())

    # Calculate percentages for each category
    catch_percentages = {cat: (count / total_catch_trials) * 100 if total_catch_trials > 0 else 0 for cat, count in catch_counts.items()}
    target_percentages = {cat: (count / total_target_trials) * 100 if total_target_trials > 0 else 0 for cat, count in target_counts.items()}
    # Plot the bar graph
    ax = data.plot(kind='bar', figsize=(8, 5), color=['tab:red', 'tab:green'], edgecolor='black')
    num_categories = len(categories)
    bar_index = 0
    for trial_type in ['Catch', 'Target']:
        for category in categories:
            bar = ax.patches[bar_index]
            height = bar.get_height()
            if height > 0:
                percentage = catch_percentages[category] if trial_type == 'Catch' else target_percentages[category]
                ax.text(
                    bar.get_x() + bar.get_width() / 2,
                    height + 1,
                    f'{percentage:.1f}%',
                    ha='center',
                    va='bottom',
                    fontsize=8
                )
            bar_index += 1

    plt.title(f'Lever Release Periods by Category for {label}')
    plt.ylabel('Count')
    plt.xlabel('Lever was released during ___ Phase')
    plt.xticks(rotation=30)
    plt.legend(title='Trial Type')
    plt.tight_layout()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
    print(data)

def clean_rt(val):
    # Handle if it's a tuple like ('cti', 1.572)
    if isinstance(val, tuple):
        if 'cti' in val or val[0] == 'cti':
            return -0.1
        return np.nan

    # Handle if it's a list or array of tuples (e.g., [('cti', 1.572), nan])
    if isinstance(val, list):
        for item in val:
            if isinstance(item, tuple) and ('cti' in item or item[0] == 'cti'):
                return -0.1
        return np.nan
    
    if isinstance(val, str) and ',nan' in val:
        return -0.1

    # Handle strings that look like tuples
    if isinstance(val, str):
        val = val.strip()
        if 'cti' in val:
            return -0.1
        elif 'fix' in val:
            return -0.3
        elif 'bar' in val:
            return -0.2
        try:
            return float(val)
        except ValueError:
            return np.nan

    # Handle regular numeric values
    try:
        return float(val)
    except:
        return np.nan
import re
def extract_value_from_rt(val):
    #note: currently some rt values are strings like "('cti', 1.572)" or "('bar', 0.5)", and they don't have a cti value because sometimes the trial ends before cti value is stored. 
    #But we still want to know when the early release happened
    if isinstance(val, str):
        match = re.match(r"\('(?:cti|bar|fix)',\s*([0-9.]+)\)", val)
        if match:
            return float(match.group(1))
    return None

def plot_rt_histogram(df, label='Dataset', bins=30):
    """
    Plots a histogram of reaction times (RT) for all 'target' trials (no CTI grouping).

    Parameters:
    - df (pd.DataFrame): DataFrame containing 'trial_type' and 'rt' columns.
    - label (str): Optional label to identify the dataset in the plot title.
    - bins (int): Number of bins for the histogram.

    Returns:
    - None
    """
    # Extract RT values and clean them
    df['clean_rt'] = df['rt'].apply(clean_rt)
    df['clean_rt'].fillna(1, inplace=True)

    # Filter for target trials only
    target_trials = df[df['trial_type'] == 'target']

    # Plot histogram
    plt.figure(figsize=(9, 6))
    plt.hist(target_trials['clean_rt'], bins=bins, color='skyblue', alpha=0.7, edgecolor='black')
    plt.xlabel('Reaction Time (RT)')
    plt.ylabel('Frequency')

    # Add reference lines
    plt.axvline(x=-0.1, color='red', linestyle='--',linewidth=0.8)
    plt.text(-0.1, plt.ylim()[1] * 1.01, 'Released during CTI', rotation=20, color='red', ha='left', fontsize=10)
    plt.axvline(x=-0.2, color='red', linestyle='--',linewidth=0.8)
    plt.text(-0.2, plt.ylim()[1] * 1.01, 'Released during Bar Phase', rotation=20, color='red', ha='left', fontsize=10)
    plt.axvline(x=-0.3, color='red', linestyle='--',linewidth=0.8)
    plt.text(-0.3, plt.ylim()[1] * 1.01, 'Released during Fixation Phase', rotation=20, color='red', ha='left', fontsize=10)
    plt.axvline(x=0.975, color='purple', linestyle='--',linewidth=0.8)
    plt.text(0.975, plt.ylim()[1] * 1.01, 'Miss', color='purple', rotation=20, ha='left', fontsize=10)
    plt.axvline(x=0, color='black', linestyle='-', linewidth=0.8)
    plt.text(0, plt.ylim()[1] * 1.01, 'Target', color='black', rotation=20, ha='left', fontsize=10)
    plt.axvline(x=0.2, color='green', linestyle='--',linewidth=0.8)
    plt.axvline(x=0.8, color='green', linestyle='--',linewidth=0.8)
    plt.text(0.2, plt.ylim()[1] * 1.01, 'Earliest Correct Response Time', rotation=20, color='green', ha='left', fontsize=10)
    plt.text(0.8, plt.ylim()[1] * 1.01, 'Latest Correct Response Time', rotation=20, color='green', ha='left', fontsize=10)

    # Final formatting
    plt.title(f'Reaction Time Histogram (All Target Trials) - {label}', loc='center', y=-0.25)
    plt.tight_layout()
    plt.show()
    print(f"Total target trials plotted: {len(target_trials)}")

def plot_rt_histogram_by_cti(df, label='Dataset', bins=30,low=0.95, high=0.95):

    """
    Plots a histogram of reaction times (RT) for 'target' trials, separated by early and late CTI.

    Parameters:
    - df (pd.DataFrame): DataFrame containing 'trial_type', 'rt', and 'cti' columns.
    - label (str): Optional label to identify the dataset in the plot title.
    - bins (int): Number of bins for the histogram.

    Returns:
    - None
    """
    df['early_release_target_cti'] = df['rt'].apply(extract_value_from_rt)
    df['clean_rt'] = df['rt'].apply(clean_rt)
    df['clean_rt'].fillna(1, inplace=True)

    # Filter for 'target' trials and remove NaNs
    early_target_trials = df.loc[(df['trial_type'] == 'target') & ((df['cti'] <= low)|(df['early_release_target_cti'] <=low))]
    late_target_trials = df.loc[(df['trial_type'] == 'target') & ((df['cti'] > high)|(df['early_release_target_cti'] > high))]

    # Plot histograms for early and late CTI
    plt.figure(figsize=(9, 6))
    plt.hist(early_target_trials['clean_rt'], bins=bins, color='blue', alpha=0.35, label=f'Early CTI (<= {low}s) (n={len(early_target_trials)})')
    plt.hist(late_target_trials['clean_rt'], bins=bins, color='orange', alpha=0.5, label=f'Late CTI (> {high}s) (n={len(late_target_trials)})')
    plt.xlabel('Reaction Time (RT)')
    plt.ylabel('Frequency')
    plt.xlim(-0.15,1.15)


    plt.axvline(x=-0.1, color='red', linestyle='--', linewidth=0.8)
    plt.text(-0.1, plt.ylim()[0] -4, 'Released \n during CTI', rotation=0, color='red', ha='center', fontsize=10)
    plt.axvline(x=0.975, color='purple', linestyle='--', linewidth=0.8)
    plt.text(0.975, plt.ylim()[0] -3, 'Miss', color='purple', rotation=0, ha='right', fontsize=10)
    plt.axvline(x=0, color='black', linestyle='-', linewidth=0.8)
    plt.text(0, plt.ylim()[1] * 1.01, 'Target', color='black', rotation=20, ha='left', fontsize=10)
    plt.axvline(x=0.2, color='green', linestyle='-', linewidth=0.8)
    plt.axvline(x=0.8, color='green', linestyle='-', linewidth=0.8)
    plt.text(0.2, plt.ylim()[1] * 1.01, 'Earliest Correct Response Time', rotation=20 ,color='green', ha='left', fontsize=10)
    plt.text(0.8, plt.ylim()[1] * 1.01, 'Latest Correct Response Time', rotation=20, color='green', ha='left', fontsize=10)

    plt.title(f'Reaction Time Histogram by CTI (Target Trials Only) {label}',loc='center', y=-0.25)
    plt.legend()
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1),fontsize=9)
    plt.tight_layout()
    plt.show()

def calculate_combined_rt_cti(df,bins = 30):
    """
    Adds a new column to the DataFrame that calculates the sum of 'clean_rt' and 'cti' columns.
    If 'clean_rt' < 0, the value from 'early_release_target_cti' is used instead.
    If 'clean_rt' < 0 and 'early_release_target_cti' is missing, the value is set to 1.6.
    If 'clean_rt' >= 1, the value is set to 1.

    Additionally, plots the distribution of 'combined_rt_cti' for correct target trials, 
    incorrect target trials, correct catch trials, and incorrect catch trials.

    Parameters:
    - df (pd.DataFrame): DataFrame containing 'clean_rt', 'cti', and 'early_release_target_cti' columns.

    Returns:
    - pd.DataFrame: The updated DataFrame with a new column 'combined_rt_cti'.
    """
    # Calculate the combined_rt_cti column
    df['combined_rt_cti'] = df.apply(
        lambda row: (
            1.6 if row['clean_rt'] < 0 and pd.isna(row['early_release_target_cti']) else
            row['early_release_target_cti'] if row['clean_rt'] < 0 else
            3 if row['clean_rt'] >= 1 else
            row['clean_rt'] + row['cti']
        ),
        axis=1
    )

    # Filter data for plotting
    correct_target = df[(df['trial_type'] == 'target') & (df['acc'] == 1)]
    incorrect_target = df[(df['trial_type'] == 'target') & (df['acc'] == 0)]
    correct_catch = df[(df['trial_type'] == 'catch') & (df['acc'] == 1)]
    incorrect_catch = df[(df['trial_type'] == 'catch') & (df['acc'] == 0)]

    # Plot the distributions
    plt.figure(figsize=(10, 6))
    alpha = 0.6
    plt.hist(correct_target['combined_rt_cti'], bins=bins, alpha=alpha, label='Correct Target', color='lightgreen')
    plt.hist(incorrect_target['combined_rt_cti'], bins=bins, alpha=alpha, label='Incorrect Target', color='darkgreen')
    plt.hist(correct_catch['combined_rt_cti'], bins=bins, alpha=alpha, label='Correct Catch', color='lightcoral')
    plt.hist(incorrect_catch['combined_rt_cti'], bins=bins, alpha=alpha, label='Incorrect Catch', color='darkred')

    # Add labels, title, and legend
    plt.xlabel('Combined RT + CTI')
    plt.ylabel('Frequency')
    plt.xlim(-0.15, 3.1)
    plt.axvline(x=np.max(incorrect_catch['combined_rt_cti']), color='tab:red', linestyle='-', linewidth=0.8)
    plt.text(np.max(incorrect_catch['combined_rt_cti']), plt.ylim()[1] * 0.5, 'Catch Trial Ending Timepoint', rotation=90, color='tab:red', ha='left', fontsize=10)
    plt.axvline(x=np.max(correct_target['cti']), color='tab:green', linestyle='-', linewidth=0.8)
    plt.text(np.max(correct_target['cti']), plt.ylim()[1] * 0.5, 'Latest Target Onset', rotation=90, color='tab:green', ha='right', fontsize=10)
    plt.axvline(x=np.min(correct_target['cti']), color='tab:green', linestyle='-', linewidth=0.8)
    plt.text(np.min(correct_target['cti']), plt.ylim()[1] * 0.5, 'Earliest Target Onset', rotation=90, color='tab:green', ha='left', fontsize=10)
    plt.axvline(x=0, color='black', linestyle='-', linewidth=0.8)
    plt.text(0, plt.ylim()[1] * 0.5, 'Cue', color='black', rotation=90, ha='left', fontsize=10)
    plt.text(3, -3.5, 'Held Lever/Missed', color='black', rotation=0, ha='center', fontsize=10)
    plt.title('Distribution of Combined RT + CTI by Trial Type and Accuracy', loc='center', y=-0.25)
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.xticks(np.arange(0, 3, 0.5))
    plt.tight_layout()
    plt.show()

    return df
```

```python
def accuracy_moving_window(df, label, window_size=50, step_size=2): 
    """
    Calculates and plots moving window accuracy for multiple trial types.

    Parameters:
    - df (pd.DataFrame): DataFrame containing 'trial_type' and 'acc' columns.
    - window_size (int): Size of the moving window.
    - step_size (int): Step size for the moving window.

    Returns:
    - None
    """
    # Initialize dictionaries to store moving accuracies for each trial type
    trial_types = df['trial_type'].unique()
    moving_accuracies = {trial_type: [] for trial_type in trial_types}
    false_alarm_rate = []

    # Iterate through the DataFrame using a sliding window
    for i in range(0, len(df) - window_size + 1, step_size):
        window = df.iloc[i:i + window_size]
        for trial_type in trial_types:
            trials = window[window['trial_type'] == trial_type]
            if len(trials) > 0:
                moving_accuracies[trial_type].append(trials['acc'].mean() * 100)
            else:
                moving_accuracies[trial_type].append(None)

        # Calculate false alarm rate for the window
        incorrect_catch_trials = window[(window['trial_type'] == 'catch') & (window['acc'] == 0)]
        early_release_target_trials = window[(window['trial_type'].isin(['valid', 'invalid_same_bar', 'invalid_adjacent_other_bar'])) & (window['acc'] == 0) & 
                                         (window['rt'].str.contains('cti|fix|bar', na=False))|(pd.to_numeric(window['rt'], errors='coerce') < 0.2)]
        total_trials_in_window = len(window)
        if total_trials_in_window > 0:
            false_alarm_rate.append((len(incorrect_catch_trials) + len(early_release_target_trials)) / total_trials_in_window * 100)
        else:
            false_alarm_rate.append(None)

    # Calculate overall accuracies for each trial type
    overall_accuracies = {
        trial_type: df[df['trial_type'] == trial_type]['acc'].mean() * 100
        for trial_type in trial_types
    }

    # Plot the moving accuracies
    plt.figure(figsize=(10, 6))
    for trial_type, accuracies in moving_accuracies.items():
        color = 'red' if trial_type == 'catch' else 'green' if trial_type == 'valid' else 'blue' if trial_type == 'invalid_same_bar' else 'orange'
        plt.plot(
            range(0, len(accuracies) * step_size, step_size),
            accuracies,
            label=f"{trial_type.capitalize()} Accuracy (Overall: {overall_accuracies[trial_type]:.2f}%)",
            alpha=0.7,
            color=color
        )

    # Plot the false alarm rate
    plt.plot(
        range(0, len(false_alarm_rate) * step_size, step_size),
        false_alarm_rate,
        label="False Alarm Rate",
        color='purple',
        alpha=0.7
    )

    # Add labels, title, and legend
    plt.ylim(0, 101)
    plt.yticks(np.arange(0, 101, 10))
    plt.xlabel('Window Start Index')
    plt.ylabel('Accuracy (%)')
    plt.title(f'Moving Window Accuracy (Window Size: {window_size}, Step Size: {step_size}) {label}')
    plt.legend()
    plt.grid(visible=True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
    plt.tight_layout()
    plt.show()

    # Calculate counts for each trial type
    trial_type_order = ['valid', 'invalid_same_bar', 'invalid_adjacent_other_bar', 'catch']
    trial_type_counts = df['trial_type'].value_counts()

    # Create a bar graph for overall accuracies
    plt.figure(figsize=(10, 6))
    bar_positions = range(len(trial_type_order) + 1)  # +1 for FA
    bar_colors = ['green', 'blue', 'orange', 'red', 'purple']
    overall_accuracies_ordered = [overall_accuracies.get(trial_type, 0) for trial_type in trial_type_order]
    overall_accuracies_ordered.append(np.mean(false_alarm_rate))  # Add FA rate

    bars = plt.bar(bar_positions, overall_accuracies_ordered, color=bar_colors, alpha=0.7, edgecolor='black')

    # Add n=... on top of the bars
    for bar, trial_type in zip(bars, trial_type_order + ['FA']):
        height = bar.get_height()
        count = trial_type_counts.get(trial_type, 0) if trial_type != 'FA' else np.mean(false_alarm_rate)
        plt.text(bar.get_x() + bar.get_width() / 2, height + 2, f'n={count}', ha='center', va='bottom', fontsize=10)

    # Add labels, title, and legend
    plt.xticks(bar_positions, [trial_type.capitalize() for trial_type in trial_type_order] + ['False Alarm'], rotation=45, ha='right')
    plt.ylabel('Accuracy (%)')
    plt.title(f'Overall Accuracy by Trial Type {label}')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Add legend
    plt.legend()
    plt.tight_layout()
    plt.show()

        # --- HIT RATE + FA GRAPH (with catch showing accuracy) ---

    trial_type_order = ['valid', 'invalid_same_bar', 'invalid_adjacent_other_bar']
    hit_rates = {}

    # Compute hit rates (excluding early releases)
    for trial_type in trial_type_order:
        trials = df[df['trial_type'] == trial_type]
        early_releases = trials[
            (trials['acc'] == 0) & (
                trials['rt'].astype(str).str.contains('cti|fix|bar', na=False) |
                (pd.to_numeric(trials['rt'], errors='coerce') < 0.2)
            )
        ]
        effective_trials = trials.drop(early_releases.index)
        if len(effective_trials) > 0:
            hit_rate = effective_trials['acc'].mean() * 100
        else:
            hit_rate = np.nan
        hit_rates[trial_type] = hit_rate

    # Add catch trial accuracy
    catch_trials = df[df['trial_type'] == 'catch']
    catch_accuracy = catch_trials['acc'].mean() * 100 if len(catch_trials) > 0 else np.nan
    hit_rates['catch'] = catch_accuracy

    # Add FA rate
    fa_rate = np.mean(false_alarm_rate)
    hit_rates['FA'] = fa_rate

    # Plot bar graph
    plt.figure(figsize=(10, 6))
    all_labels = ['Valid', 'Invalid_same_bar', 'Invalid_adjacent_other_bar', 'Catch (Accuracy)', 'False Alarm']
    bar_colors = ['green', 'blue', 'orange', 'red', 'purple']
    hit_rate_values = [hit_rates.get(tt, 0) for tt in ['valid', 'invalid_same_bar', 'invalid_adjacent_other_bar', 'catch', 'FA']]

    bars = plt.bar(range(len(hit_rate_values)), hit_rate_values, color=bar_colors, alpha=0.7, edgecolor='black')

    # Add n=... or % label on top of bars
    for bar, trial_type in zip(bars, ['valid', 'invalid_same_bar', 'invalid_adjacent_other_bar', 'catch', 'FA']):
        height = bar.get_height()
        if trial_type == 'FA':
            labels = f'{height:.1f}%'
        else:
            n = df[df['trial_type'] == trial_type].shape[0]
            labels = f'n={n}'
        plt.text(bar.get_x() + bar.get_width() / 2, height + 2, labels, ha='center', va='bottom', fontsize=10)

    plt.xticks(range(len(hit_rate_values)), all_labels, rotation=45, ha='right')
    plt.ylabel('Hit Rate / Accuracy (%)')
    plt.title(f'Hit Rate by Trial Type (Excludes Early Releases) {label}')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.ylim(0, 101)
    plt.tight_layout()
    plt.show()
    
def plot_cti_vs_accuracy(df, label='Dataset', bins=30):
    """
    Plots accuracy vs CTI for different trial types (valid, invalid_same_bar, invalid_adjacent_other_bar).

    Parameters:
    - df (pd.DataFrame): DataFrame containing 'trial_type', 'cti', and 'acc' columns.
    - label (str): Optional label to identify the dataset in the plot title.
    - bins (int): Number of bins for the histogram.

    Returns:
    - None
    """
    # Filter for trials with valid CTI and accuracy values
    df = df.loc[df['cti'].notna() & df['acc'].notna()].copy()

    # Define CTI bins
    num_bins = bins
    bins = np.linspace(df['cti'].min(), df['cti'].max(), num_bins + 1)

    # Assign each CTI to a bin
    df['cti_bin'] = pd.cut(df['cti'], bins=bins, include_lowest=True)

    # Initialize the plot
    plt.figure(figsize=(10, 6))

    # Define trial types and colors
    trial_types = ['valid', 'invalid_same_bar', 'invalid_adjacent_other_bar']
    colors = ['green', 'orange', 'blue']

    # Plot accuracy for each trial type
    for trial_type, color in zip(trial_types, colors):
        trial_df = df[df['trial_type'] == trial_type]
        if not trial_df.empty:
            accuracies = trial_df.groupby('cti_bin', observed=False)['acc'].mean()
            bin_centers = [interval.mid for interval in accuracies.index]
            plt.plot(bin_centers, accuracies.values, marker='o', label=f'{trial_type.capitalize()} Trials', color=color)

    # Add labels, title, and legend
    plt.xlabel('Cue-to-Target Interval (CTI)')
    plt.ylabel('Accuracy')
    plt.title(f'Accuracy vs CTI for Different Trial Types ({label})')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def _coerce_numeric_cti(df, cti_col="cti", drop_nan=True):
    out = df.copy()
    out[cti_col] = pd.to_numeric(out[cti_col], errors="coerce")
    if drop_nan:
        out = out.dropna(subset=[cti_col])
    return out

def _map_trial_types(df, src_col="trial_type", dst_col="analysis_type"):
    mapping = {
        "valid": "cued",
        "invalid_same_bar": "same_object",
        "invalid_adjacent_other_bar": "different_object",
    }
    out = df.copy()
    out[dst_col] = out[src_col].map(mapping)
    out = out[~out[dst_col].isna()]
    return out

def analyze_cti_oscillations_by_type(
    df,
    *,
    label="Dataset",
    trial_col="analysis_type",
    types=None,
    max_cti=1.7,
    cti_min=0.30,
    bin_size=None,
    win_size=0.05,
    step=0.01,
    use_sliding=True,
    detrend=True,
    detrend_order=1,
    do_fft=True,
    fmin=2.0, fmax=20.0,
    difference_pair=("same_object","different_object"),
    return_results=True
):
    needed = {"cti", "acc", trial_col}
    if not needed.issubset(df.columns):
        missing = ", ".join(sorted(needed - set(df.columns)))
        raise ValueError(f"DataFrame missing required columns: {missing}")

    # colors
    COLORS = {"cued": "black", "same_object": "orange", "different_object": "royalblue"}

    dff = df.copy()
    dff = dff[(dff["cti"] > 0) & (dff["cti"] < max_cti)]
    if cti_min is not None:
        dff = dff[dff["cti"] >= float(cti_min)]
    dff = dff.dropna(subset=["cti", "acc"])
    if dff.empty:
        raise ValueError("No rows left after CTI filtering. Check cti_min/max_cti and data.")

    if types is None:
        types = pd.Index(dff[trial_col].dropna().unique()).tolist()

    lo, hi = dff["cti"].min(), dff["cti"].max()

    if use_sliding:
        centers = np.arange(lo + win_size/2, hi - win_size/2 + 1e-9, step)
        windows = [(c - win_size/2, c + win_size/2) for c in centers]
        def build_series(sub):
            y = np.full(len(centers), np.nan, float)
            cti_vals = sub["cti"].to_numpy(); acc_vals = sub["acc"].to_numpy()
            for i, (a, b) in enumerate(windows):
                m = (cti_vals >= a) & (cti_vals < b)
                if m.any():
                    y[i] = np.nanmean(acc_vals[m])
            return centers, y
        time_axis = centers
        effective_dt = step
        mode_str = "sliding"
    else:
        if bin_size is None:
            bin_size = 0.05
        bins = np.arange(lo, hi + bin_size, bin_size)
        bin_centers = (bins[:-1] + bins[1:]) / 2
        def build_series(sub):
            s = pd.cut(sub["cti"], bins=bins, include_lowest=True)
            y = sub.groupby(s)["acc"].mean()
            cat = pd.cut(pd.Series(bins[:-1]), bins=bins, include_lowest=True).cat.categories
            y = y.reindex(cat).to_numpy()
            return bin_centers, y
        time_axis = bin_centers
        effective_dt = bin_size
        mode_str = f"fixed ({int(bin_size*1000)} ms)"

    # per-type series (raw & detrended)
    ts = {}
    for tname in types:
        sub = dff[dff[trial_col] == tname]
        if sub.empty:
            continue
        t, y = build_series(sub)
        if detrend:
            ok = np.isfinite(y)
            if ok.sum() >= (detrend_order + 1):
                coeffs = np.polyfit(t[ok], y[ok], deg=detrend_order)
                trend = np.polyval(coeffs, t)
                y_det = y - trend
            else:
                y_det = y - np.nanmean(y)
        else:
            y_det = y.copy()
        ts[tname] = (t, y, y_det)

    # === PLOT 1: RAW time series ===
    plt.figure(figsize=(8, 5))
    for tname, (t, y, _) in ts.items():
        c = COLORS.get(tname)
        plt.plot(t, y * 100.0, lw=2.0, color=c, label=tname)
    plt.title(f"CTI vs Accuracy (RAW) by {trial_col} — {label} [{mode_str}]")
    plt.xlabel("CTI (s)"); plt.ylabel("Hit Rate (%)")
    plt.grid(True, alpha=0.3); plt.legend()
    plt.tight_layout(); plt.show()

    # === PLOT 2: DETRENDED time series ===
    plt.figure(figsize=(8, 5))
    for tname, (t, _, y_det) in ts.items():
        c = COLORS.get(tname)
        plt.plot(t, y_det * 100.0, lw=2.0, color=c, label=tname)
    plt.title(f"CTI vs Accuracy (DETRENDED) by {trial_col} — {label} [{mode_str}]")
    plt.xlabel("CTI (s)"); plt.ylabel("Hit Rate (%)")
    plt.grid(True, alpha=0.3); plt.legend()
    plt.tight_layout(); plt.show()

    # Difference wave (detrended): same − different
    diff_result = None
    if difference_pair is not None and all(t in ts for t in difference_pair):
        A, B = difference_pair
        t = ts[A][0]
        yA = ts[A][2]; yB = ts[B][2]
        n = min(len(yA), len(yB))
        diff = yA[:n] - yB[:n]
        plt.figure(figsize=(8, 4))
        plt.plot(t[:n], diff * 100.0, lw=2.0, color="dimgray")
        plt.axhline(0, ls="--", alpha=0.6, color="gray")
        plt.title(f"Difference wave (DETRENDED): {A} − {B}")
        plt.xlabel("CTI (s)"); plt.ylabel("Δ Hit Rate (%)")
        plt.grid(True, alpha=0.3); plt.tight_layout(); plt.show()
        diff_result = (t[:n], diff)

    # Spectra (on detrended series)
    spectra = {}
    if do_fft:
        def _fft_series(y, dt):
            y = y - np.nanmean(y)
            y = np.where(np.isfinite(y), y, 0.0)
            w = np.hanning(len(y))
            Y = np.fft.rfft(y * w)
            freqs = np.fft.rfftfreq(len(y), d=dt)
            amp = np.abs(Y)
            return freqs, amp

        plt.figure(figsize=(10, 5))
        for tname, (_, _, y_det) in ts.items():
            freqs, amp = _fft_series(y_det, effective_dt)
            m = (freqs >= fmin) & (freqs <= fmax)
            plt.plot(freqs[m], amp[m], lw=2.0, color=COLORS.get(tname), label=tname)
            spectra[tname] = {"freqs": freqs, "amp": amp}
        plt.title(f"Behavioral spectrum (DETRENDED) — {label}")
        plt.xlabel("Frequency (Hz)"); plt.ylabel("Spectral amplitude (a.u.)")
        plt.grid(True, alpha=0.3); plt.legend(); plt.tight_layout(); plt.show()

        if diff_result is not None:
            t, diff = diff_result
            freqs, amp = _fft_series(diff, effective_dt)
            m = (freqs >= fmin) & (freqs <= fmax)
            plt.figure(figsize=(10, 4))
            plt.plot(freqs[m], amp[m], lw=2.0, color="dimgray")
            plt.title(f"Spectrum of difference (DETRENDED): {difference_pair[0]} − {difference_pair[1]}")
            plt.xlabel("Frequency (Hz)"); plt.ylabel("Spectral amplitude (a.u.)")
            plt.grid(True, alpha=0.3); plt.tight_layout(); plt.show()

    if return_results:
        return {
            "time_axis": time_axis,
            "time_series": ts,
            "spectra": spectra,
            "difference": diff_result,
            "dt": effective_dt
        }

# -----------------------
# EXAMPLE USAGE ON YOUR DATAFRAME `df`
# -----------------------
df=master_df
# 1) Clean CTI (handles entries like ('cti', 0.529))
df_clean = _coerce_numeric_cti(df, cti_col="cti")

# 2) Map trial_type to Egly-style analysis buckets; drops 'catch'
df_egly = _map_trial_types(df_clean, src_col="trial_type", dst_col="analysis_type")

# 3) Run analysis (sliding 50ms windows, 10ms step, 0.30–1.70s)
res = analyze_cti_oscillations_by_type(
    df_egly,
    label="Subject / Session",
    trial_col="analysis_type",
    types=["cued","same_object","different_object"],
    max_cti=1.69,
    cti_min=0.30,
    use_sliding=True,     # like the paper: 50 ms windows, 10 ms step
    win_size=0.05,
    step=0.01,
    detrend=True,
    detrend_order=1,
    do_fft=True,
    fmin=2.0, fmax=20.0,
    difference_pair=("same_object","different_object")
)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
def calculate_stats_table_extended(datasets, labels):
    """
    Compute accuracy and false‑alarm stats for datasets that include
    'valid', 'invalid_same_bar', 'invalid_adjacent_other_bar', 'catch'.

    Returns
    -------
    pd.DataFrame  # one row per dataset
    """
    if len(datasets) != len(labels):
        raise ValueError("datasets and labels must be same length")

    trial_types = ['valid', 'invalid_same_bar',
                   'invalid_adjacent_other_bar', 'catch']
    rows = []

    for df, lab in zip(datasets, labels):
        if not {'trial_type', 'acc', 'rt'}.issubset(df.columns):
            raise ValueError("Each DataFrame must contain 'trial_type', 'acc', 'rt'.")

        row = {'Dataset': lab, 'Total Trials': len(df)}

        # ---- accuracy per trial type with counts ----
        for tt in trial_types:
            trials = df[df['trial_type'] == tt]
            pct   = trials['acc'].mean() * 100 if len(trials) else np.nan
            row[f'{tt}_acc_%'] = f"{pct:.2f} ({trials['acc'].sum()}/{len(trials)})"

        # ---- false‑alarm rate (same rule) ----
        incorrect_catch = df[(df['trial_type'] == 'catch') & (df['acc'] == 0)]
        early_release   = df[
            (df['trial_type'].isin(['valid', 'invalid_same_bar', 'invalid_adjacent_other_bar']))
            & (df['acc'] == 0)
            & (
                df['rt'].astype(str).str.contains('cti|fix|bar', na=False)
                | (pd.to_numeric(df['rt'], errors='coerce') < 0.2)
            )
        ]
        fa_rate = ((len(incorrect_catch) + len(early_release)) / len(df) * 100
                   if len(df) else np.nan)
        row['fa_rate_%'] = f"{fa_rate:.2f} (({len(early_release)}+{len(incorrect_catch)})/{len(df)})"

        rows.append(row)

    return pd.DataFrame(rows)

def plot_stats_table_extended(datasets, labels):
    """
    Summarise and plot performance metrics for tasks that contain
    the trial types: valid, invalid_same_bar, invalid_adjacent_other_bar, catch.

    Parameters
    ----------
    datasets : list[pd.DataFrame]
        Each DataFrame must contain at least the columns
        'trial_type', 'acc', and 'rt'.
    labels : list[str]
        One label per DataFrame.

    Returns
    -------
    stats_df : pd.DataFrame
        One row per dataset with accuracy for every trial type
        and the overall false‑alarm rate.
    """
    if len(datasets) != len(labels):
        raise ValueError("The number of datasets must match the number of labels.")

    trial_types = ['valid', 'invalid_same_bar',
                   'invalid_adjacent_other_bar', 'catch']
    colour_map = {
        'valid'                     : 'green',
        'invalid_same_bar'          : 'blue',
        'invalid_adjacent_other_bar': 'orange',
        'catch'                     : 'red',
        'fa_rate'                   : 'purple'
    }

    stats = []

    for df, label in zip(datasets, labels):

        if not {'trial_type', 'acc', 'rt'}.issubset(df.columns):
            raise ValueError("Each DataFrame must contain 'trial_type', 'acc', and 'rt' columns.")

        row = {'Dataset': label}

        # -------- accuracy per trial type --------
        for tt in trial_types:
            trials = df[df['trial_type'] == tt]
            row[f'{tt}_acc_%'] = trials['acc'].mean() * 100 if len(trials) else np.nan

        # -------- false‑alarm rate (same rule you used downstream) --------
        incorrect_catch = df[(df['trial_type'] == 'catch') & (df['acc'] == 0)]
        early_release   = df[
            (df['trial_type'].isin(['valid', 'invalid_same_bar', 'invalid_adjacent_other_bar']))
            & (df['acc'] == 0)
            & (
                df['rt'].astype(str).str.contains('cti|fix|bar', na=False)
                | (pd.to_numeric(df['rt'], errors='coerce') < 0.2)
            )
        ]
        row['fa_rate_%'] = (
            (len(incorrect_catch) + len(early_release)) / len(df) * 100
            if len(df) else np.nan
        )

        stats.append(row)

    stats_df = pd.DataFrame(stats)

    # ------------------------------ PLOT ------------------------------
    fig, ax = plt.subplots(figsize=(12, 6))
    x = np.arange(len(stats_df))

    # -----------------------------------------
    # inside plot_stats_table_extended
    # -----------------------------------------
    for metric in [f'{tt}_acc_%' for tt in trial_types] + ['fa_rate_%']:

        # -------- pick the right colour key --------
        if 'invalid_same_bar' in metric:
            colour_key = 'invalid_same_bar'
        elif 'invalid_adjacent_other_bar' in metric:
            colour_key = 'invalid_adjacent_other_bar'
        elif 'valid' in metric:
            colour_key = 'valid'
        elif 'catch' in metric:
            colour_key = 'catch'
        else:                      # fa_rate_%
            colour_key = 'fa_rate'

        # -------- build a pretty label on the fly --------
        pretty_label = (
            metric.replace('_acc_%', ' Accuracy (%)')
                .replace('valid', 'Valid')
                .replace('invalid_same_bar', 'Invalid Same‑bar')
                .replace('invalid_adjacent_other_bar', 'Invalid Adjacent Other‑bar')
                .replace('catch', 'Catch')
                .replace('fa_rate_%', 'False‑Alarm Rate (%)')
        )

        ax.plot(
            x, stats_df[metric],
            marker='o', linestyle='-',
            alpha=0.8,
            label=pretty_label,
            color=colour_map[colour_key]
        )



    ax.set_xticks(x)
    ax.set_xticklabels(stats_df['Dataset'], rotation=45, ha='right', fontsize=10)
    ax.set_yticks(np.arange(0, 101, 10))
    ax.set_ylim(0, 101)

    ax.set_xlabel('Dataset', fontsize=12)
    ax.set_ylabel('Percentage (%)', fontsize=12)
    ax.set_title('Performance Metrics by Dataset', fontsize=14)
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    ax.legend(fontsize=9)
    plt.tight_layout()
    plt.show()

    return stats_df

datasets = [j20, j24, j26, j28, j30, ju01, ju03, ju07, ju09, ju11, ju14, ju16, ju18,ju21,ju23,ju25, ju30,ju31, au01,au05, au07,au08,au11,au13,au14,au18,au20,au22,au26,au28,au29,s02,s03,s06,s08,s11,s12,s16,s19 ]
labels = ['06/20', '06/24', '06/26', '06/28', '06/30', '07/01', '07/03', '07/07', '07/09', '07/11', '07/14', '07/16', '07/18','07/21','07/23','07/25','07/30','07/31','08/01','08/05','08/07','08/08','08/11','08/13','08/14','08/18','08/20','08/22','08/26','08/28','08/29','9/2','9/3','9/6','9/8','9/11','9/12','9/16','9/19']
#calculate_stats_table(datasets, labels)
plot_stats_table_extended(datasets, labels)
```

```python
# Re-import libraries after code execution state reset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def compute_hit_rate(df, trial_type):
    """Compute hit rate excluding early releases."""
    trials = df[df['trial_type'] == trial_type]
    early_releases = trials[
        (trials['acc'] == 0) & (
            trials['rt'].astype(str).str.contains('cti|fix|bar', na=False) |
            (pd.to_numeric(trials['rt'], errors='coerce') < 0.2)
        )
    ]
    effective_trials = trials.drop(early_releases.index)
    if len(effective_trials) > 0:
        return effective_trials['acc'].mean() * 100
    else:
        return np.nan

def plot_hit_rate_table(datasets, labels):
    """
    Plot hit rates for valid, invalid_same_bar, invalid_adjacent_other_bar,
    and catch accuracy + FA rate, excluding early releases.
    """
    trial_type_order = ['valid', 'invalid_same_bar', 'invalid_adjacent_other_bar']
    colour_map = {
        'valid': 'green',
        'invalid_same_bar': 'blue',
        'invalid_adjacent_other_bar': 'orange',
        'catch': 'red',
        'FA': 'purple'
    }

    rows = []
    for df, label in zip(datasets, labels):
        row = {'Dataset': label}
        for tt in trial_type_order:
            row[tt] = compute_hit_rate(df, tt)
        catch_trials = df[df['trial_type'] == 'catch']
        row['catch'] = catch_trials['acc'].mean() * 100 if len(catch_trials) > 0 else np.nan

        incorrect_catch = df[(df['trial_type'] == 'catch') & (df['acc'] == 0)]
        early_release = df[
            (df['trial_type'].isin(trial_type_order)) & (df['acc'] == 0) & (
                df['rt'].astype(str).str.contains('cti|fix|bar', na=False) |
                (pd.to_numeric(df['rt'], errors='coerce') < 0.2)
            )
        ]
        row['FA'] = (len(incorrect_catch) + len(early_release)) / len(df) * 100 if len(df) else np.nan
        rows.append(row)

    hit_df = pd.DataFrame(rows)

    # Plotting
    fig, ax = plt.subplots(figsize=(12, 6))
    x = np.arange(len(hit_df))

    for key in trial_type_order + ['catch', 'FA']:
        ax.plot(
            x, hit_df[key],
            marker='o', linestyle='-',
            alpha=0.8,
            label=key.replace('_', ' ').capitalize() if key != 'FA' else 'False Alarm',
            color=colour_map[key]
        )

    ax.set_xticks(x)
    ax.set_xticklabels(hit_df['Dataset'], rotation=45, ha='right', fontsize=10)
    ax.set_yticks(np.arange(0, 101, 10))
    ax.set_ylim(0, 101)
    ax.set_xlabel('Dataset', fontsize=12)
    ax.set_ylabel('Percentage (%)', fontsize=12)
    ax.set_title('Hit Rate by Trial Type (Excludes Early Releases)', fontsize=14)
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    ax.legend(fontsize=9)
    plt.tight_layout()
    plt.show()

    return hit_df
plot_hit_rate_table(datasets, labels)
```

```python
def plot_hit_rate_vs_cti(df):
    """
    Generates a plot of hit rate vs. CTI value.
    
    The function assumes that df has at least the following columns:
      - 'rt': reaction time values that can be numeric, NaN, or strings (e.g., 'nan', 'fix', 'bar', 'cti')
      - 'trial_type': should include at least 'valid', 'invalid_same_bar', 'invalid_adjacent_other_bar' for target trials.
      - 'cti': a numeric column representing the cue-target interval value.
    
    For target trials:
      - "Hits" are defined as those where the transformed RT is between 0 and 0.85.
      - "Misses" (Held Lever) are those with a transformed RT equal to 1.
      
    Hit rate is calculated as:
        hit_rate = hits / (hits + misses)
    """
    
    # Helper function to transform RT values
    def transform_val(val):
        # Treat NaN (or the string "nan") as no response (Held Lever: 1)
        if pd.isna(val) or (isinstance(val, str) and val.lower() == 'nan'):
            return 1
        # Special phase labels: 'fix', 'bar', or 'cti'
        if isinstance(val, str):
            if "fix" in val:
                return -0.3
            elif "bar" in val:
                return -0.2
            elif "cti" in val:
                return -0.1
        # Otherwise, attempt conversion to float
        try:
            return float(val)
        except (ValueError, TypeError):
            return np.nan

    # Work on a copy so the original DataFrame is not modified
    df = df.copy()
    df['rt_special'] = df['rt'].apply(transform_val)
    
    # Focus only on target trials (valid, invalid_same_bar, invalid_adjacent_other_bar)
    target_df = df[df['trial_type'].isin(['valid', 'invalid_same_bar', 'invalid_adjacent_other_bar'])]
    
    # Group target trials by CTI value and compute hit rates
    hit_rates = {}
    for cti_val, group in target_df.groupby('cti'):
        # Hits: RT values between 0 and 0.85 (correct responses)
        hits = group[(group['rt_special'] > 0.2) & (group['rt_special'] < 0.8)].shape[0]
        # Misses: Held lever responses (rt_special == 1)
        misses = group[group['rt_special'] == 1].shape[0]
        total = hits + misses
        hit_rate = hits / total if total > 0 else np.nan
        hit_rates[cti_val] = hit_rate
    
    # Sort CTI values for plotting
    sorted_cti = sorted(hit_rates.keys())
    sorted_hit_rates = [hit_rates[cti] for cti in sorted_cti]
    
    # Create the plot
    plt.figure(figsize=(8, 6))
    plt.plot(sorted_cti, sorted_hit_rates, marker='o', linestyle='-')
    plt.xlabel('CTI Value')
    plt.ylabel('Hit Rate')
    plt.ylim(0,1)
    plt.title('Hit Rate vs. CTI Value')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def plot_cti_bins_vs_fa_rate(df, num_bins=20):
    """
    Plots FA rate vs CTI bins.

    Parameters:
    - df (pd.DataFrame): DataFrame containing 'trial_type', 'cti', and 'acc' columns.
    - num_bins (int): Number of bins for CTI.

    Returns:
    - None
    """
    # Filter for trials with valid CTI values
    filtered_df = df[df['cti'].notna()].copy()

    # Define CTI bins
    bins = np.linspace(filtered_df['cti'].min(), filtered_df['cti'].max(), num_bins + 1)

    # Assign each CTI to a bin
    filtered_df['cti_bin'] = pd.cut(filtered_df['cti'], bins=bins, include_lowest=True)

    # Calculate FA rate for each CTI bin
    fa_rates = []
    bin_centers = []
    for cti_bin, group in filtered_df.groupby('cti_bin', observed=False):
        overall_target_early_release = len(group[(group['trial_type'].isin(['valid', 'invalid_same_bar', 'invalid_adjacent_other_bar'])) & (group['acc'] == 0) & ((group['rt'].str.contains('cti|fix|bar', na=False)) | (pd.to_numeric(group['rt'], errors='coerce') < 0.2))])
        overall_incorrect_catch = len(group[(group['trial_type'] == 'catch') & (group['acc'] == 0)])
        overall_fa_rate = ((overall_target_early_release + overall_incorrect_catch) / len(group)) * 100 if len(group) > 0 else 0
        fa_rates.append(overall_fa_rate)
        bin_centers.append(cti_bin.mid)

    # Plot FA rate vs CTI bins
    plt.figure(figsize=(8, 5))
    plt.plot(bin_centers, fa_rates, marker='o', color='blue', label='FA Rate')
    plt.xlabel('Cue-to-Target Interval (CTI)')
    plt.ylabel('False Alarm Rate (%)')
    plt.ylim(0,100)
    plt.title('FA Rate vs CTI Bins')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_hit_rate_vs_cti(j28)
plot_cti_bins_vs_fa_rate(j28, num_bins=20)
```

```python
def accuracy_moving_window(df, window_size=50, step_size=2): 
    """
    Calculates and plots moving window accuracy for multiple trial types.

    Parameters:
    - df (pd.DataFrame): DataFrame containing 'trial_type' and 'acc' columns.
    - window_size (int): Base size of the moving window.
    - step_size (int): Step size for the moving window.

    Returns:
    - None
    """
    # Initialize dictionaries to store moving accuracies for each trial type
    trial_types = df['trial_type'].unique()
    moving_accuracies = {trial_type: [] for trial_type in trial_types}
    false_alarm_rate = []

    # Iterate through the DataFrame using a sliding window
    for i in range(0, len(df) - window_size + 1, step_size):
        window = df.iloc[i:i + window_size]
        for trial_type in trial_types:
            # Adjust window size for invalid trials
            adjusted_window_size = window_size * 5 if trial_type in ['invalid_adjacent_other_bar', 'invalid_same_bar'] else window_size
            adjusted_window = df.iloc[i:i + adjusted_window_size]

            trials = adjusted_window[adjusted_window['trial_type'] == trial_type]
            if len(trials) > 0:
                moving_accuracies[trial_type].append(trials['acc'].mean() * 100)
            else:
                moving_accuracies[trial_type].append(None)

        # Calculate false alarm rate for the window
        incorrect_catch_trials = window[(window['trial_type'] == 'catch') & (window['acc'] == 0)]
        early_release_invalid_trials = window[(window['trial_type'].str.contains('invalid', na=False)) & (window['acc'] == 0)]
        total_trials_in_window = len(window)
        if total_trials_in_window > 0:
            false_alarm_rate.append((len(incorrect_catch_trials) + len(early_release_invalid_trials)) / total_trials_in_window * 100)
        else:
            false_alarm_rate.append(None)

    # Calculate overall accuracies for each trial type
    overall_accuracies = {
        trial_type: df[df['trial_type'] == trial_type]['acc'].mean() * 100
        for trial_type in trial_types
    }

    # Plot the moving accuracies
    plt.figure(figsize=(10, 6))
    for trial_type, accuracies in moving_accuracies.items():
        color = 'red' if trial_type == 'catch' else 'green' if trial_type == 'valid' else 'blue' if trial_type == 'invalid_same_bar' else 'orange'
        plt.plot(
            range(0, len(accuracies) * step_size, step_size),
            accuracies,
            label=f"{trial_type.capitalize()} Accuracy (Overall: {overall_accuracies[trial_type]:.2f}%)",
            alpha=0.7,
            color=color
        )

    # Plot the false alarm rate
    plt.plot(
        range(0, len(false_alarm_rate) * step_size, step_size),
        false_alarm_rate,
        label="False Alarm Rate",
        color='purple',
        alpha=0.7
    )

    # Add labels, title, and legend
    plt.ylim(0, 101)
    plt.yticks(np.arange(0, 101, 10))
    plt.xlabel('Window Start Index')
    plt.ylabel('Accuracy (%)')
    plt.title(f'Moving Window Accuracy (Window Size: {window_size}, Step Size: {step_size})')
    plt.legend()
    plt.grid(visible=True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
    plt.tight_layout()
    plt.show()
```
